#### Pairplot with Seaborn 
It is often the case when we need to pairplot some variables of a list of numpy matrix, where 

- often the column/variable/field name is missing
- pairplot is compatible with pandas

An easy way is as follows

```python
import pandas as pd
import numpy as np
import os
# import sklearn.cluster as cluster
import seaborn as sns
```

```python
# FA = cluster.FeatureAgglomeration(n_clusters=7,compute_full_tree=True, linkage='ward')
# FA.fit(data)
# X_reduced = FA.transform(data)
# Previous is how the data comes - from broad-band acoustics data with Feature Agglomerative transformation

file_path = 'data/'
file_name = 'FeatureAgglomeration.csv'

X_reduced = pd.read_csv(os.path.join(file_path, file_name))
X_reduced = X_reduced.values

# if there's no column name and index, etc.
index = ['Feature '+ str(i+1) for i in range(X_reduced.shape[1]-1)]
index.append('labels')

# number of samples to plot
n_2plot = 1000

df_X = pd.DataFrame.from_records(X_reduced[:,:], columns=index)
# df_X.to_csv('FeatureAgglomeration.csv', index=False) # do not add index column 

# print(df_X.head())


# There's a lot more to explore with pairplot
g=sns.pairplot(df_X, diag_kind="hist", hue = 'labels', palette = 'deep',plot_kws={"s":3})
g.fig.set_size_inches(20,20)

# plt.savefig(os.path.join("results", "foo.png"))
```
