#### Pairplot with Seaborn 
It is often the case when we need to pairplot some variables of a list of numpy matrix, where 

- often the column/variable/field name is missing
- pairplot is compatible with pandas

An easy way is as follows

```python
import pandas as pd
import numpy as np
# import sklearn.cluster as cluster
import seaborn as sns
```

```python
# FA = cluster.FeatureAgglomeration(n_clusters=7,compute_full_tree=True, linkage='ward')
# FA.fit(data)

# X_reduced = FA.transform(data)

X_reduced = pd.read_csv('FeatureAgglomeration.csv')
X_reduced = X_reduced.values
# if there's no column name and index, etc.
index = ['Feature '+str(i) for i in range(X_reduced.shape[1]-1)]
index.append('labels')

# number of samples to plot
n_2plot = 1000

df_X = pd.DataFrame.from_records(X_reduced[:,:], columns=index)
print(df_X.head())

g=sns.pairplot(df_X, diag_kind="hist", hue = 'labels', palette = 'colorblind',plot_kws={"s":3})
g.fig.set_size_inches(20,20)

# plt.savefig(os.path.join("results", "cluster_scatter_FeatureAgglomeration_nc%d_pc.png"%n_clusters))
```
